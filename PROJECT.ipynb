```python
# ANOVA for conduct
fvalue, pvalue = stats.f_oneway(data['Cond1'], data['Cond2'], data['Cond3'],data['Cond4'])
print(fvalue, pvalue)

# cond1 against everything else
a = data['Cond1']
b = data['Cond2']
t1 = stats.ttest_ind(a,b)
print(t1)

a = data['Cond1']
b = data['Cond3']
t2 = stats.ttest_ind(a,b)
print(t2)

a = data['Cond1']
b = data['Cond4']
t3 = stats.ttest_ind(a,b)
print(t3)

# cond2 against everything else
a = data['Cond2']
b = data['Cond3']
t4 = stats.ttest_ind(a,b)
print(t4)

a = data['Cond2']
b = data['Cond4']
t5 = stats.ttest_ind(a,b)
print(t5)

# cond3 against everything else
#only sig. one
a = data['Cond3']
b = data['Cond4']
t6 = stats.ttest_ind(a,b)
print(t6)

condata = [
    ["Gs1 vs. Gs2",t1],
    ["Gs1 vs. Gs3",t2],
    ["Gs1 vs. Gs4",t3],
    ["Gs2 vs. Gs3",t4],
    ["Gs2 vs. Gs4",t5],
    ["Gs3 vs. Gs4",t6]]

headd = ["T-test", "P-value"]

print(tabulate(condata, headers=headd, tablefmt="grid"))
```

```python
# ANOVA for photo
phfvalue, phpvalue = stats.f_oneway(data['Photo1'], data['Photo2'], data['Photo3'],data['Photo4'])
print(fvalue, pvalue)

# ANOVA for co2
cofvalue, copvalue = stats.f_oneway(data['Ci1'], data['Ci2'], data['Ci3'],data['Ci4'])
print(fvalue, pvalue)

# ANOVA for vpd
vpfvalue, vppvalue = stats.f_oneway(data['VPD1'], data['VPD2'], data['VPD3'],data['VPD4'])
print(fvalue, pvalue)
```

```python
from tabulate import tabulate

mydata = [
    ["photosynthesis",phfvalue, phpvalue],
    ["internal CO2", cofvalue,copvalue],
    ["Vapor pressure deficit",vpfvalue,vppvalue]]

# create header
head = ["variable","f-value", "p-value"]
 
# display table
print(tabulate(mydata, headers=head, tablefmt="grid"))
```
